```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
spark.sql("USE CATALOG `quiz`")
spark.sql("USE SCHEMA `source`")
```

```python
# Source Schema Implementation
# Copy data from data store to Source table in Databricks

# Read data from data store ('csv')
df = spark.read.option(
    "inferSchema", "true"
).option(
    "header", "true"
).option(
    "sep", "|"
).csv(
    "/Volumes/quiz/source/datastore/material_master_1k.csv"
)

# Write data to Databricks Table
df.write.mode("overwrite").saveAsTable("quiz.source.material_master")

```

```python
%sql
ALTER TABLE quiz.source.material_master
SET TBLPROPERTIES (delta.enableChangeDataFeed = true);
```
